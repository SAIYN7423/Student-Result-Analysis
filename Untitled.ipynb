# drop un nambe columan

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# change weekly study hour columna

```python
df["WklyStudyHours"] =df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

# gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender siatribution")
plt.show()
```

# from the above chart we have analyswd thst:
#the number of female in data is more the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (4,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's and student's score")
plt.show()
```

```python
# from the above chart we have  concluded the the eduction of the parents have a good impact on thair
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize =(4,5))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between parent's marital status and stdents's score")
plt.show()
```

```python
# from the above chat we concluded that ther is n0/neglibile impact on the 
# children's score due to prent martial status 
```

# distribuyion of ethic groupss

```python
groupA = df.loc[(df['EthnicGroup'] =="group A")].count()
groupB = df.loc[(df['EthnicGroup'] =="group B")].count()
groupC = df.loc[(df['EthnicGroup'] =="group C")].count()
groupD = df.loc[(df['EthnicGroup'] =="group D")].count()
groupE = df.loc[(df['EthnicGroup'] =="group E")].count()

l = ["group A","group B","group C","group D",'group E']
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels = l, autopct =  "%1.2f%%")
plt.title(" Distribuyion of ethic groupss")
plt.show()
```
